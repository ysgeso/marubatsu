```python
X = [0, 1]      # X がとりうる値を表す list
P = [1/2, 1/2]  # X が従う確率分布を表す list
total = 0
for x in X:
    total += x * P[x]
print(total)
```

```python
X = [0, 1]       # X がとりうる値を表す list
Y = [0, 1]       # Y がとりうる値を表す list
PX = [1/2, 1/2]  # X の確率分布を表す list
PY = [1/3, 2/3]  # Y の確率分布を表す list
total = 0
for x in X:
    for y in Y:
        total += (x + y) * PX[x] * PY[y]
print(total)
```
