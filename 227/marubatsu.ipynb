```python
import matplotlib.pyplot as plt
import japanize_matplotlib
import numpy as np

X = np.array([-2, -1, 0, 1, 2])
Y = X * 2
P = [0.1, 0.25, 0.3, 0.25, 0.1]

plt.bar(X, P)
plt.xlim(-5, 5) # X の表示範囲を -5 ~ 5 にする
plt.xlabel("X")
plt.ylabel("確率")
plt.show()
plt.bar(Y, P)
plt.xlim(-5, 5) # Y の表示範囲を -5 ~ 5 にする
plt.xlabel("Y")
plt.ylabel("確率")
plt.show()
```

```python
Z = X + 2
plt.bar(X, P)
plt.xlim(-5, 5) # X の表示範囲を -5 ~ 5 にする
plt.xlabel("X")
plt.ylabel("確率")
plt.show()
plt.bar(Z, P)
plt.xlim(-5, 5) # Z の表示範囲を -5 ~ 5 にする
plt.xlabel("Z")
plt.ylabel("確率")
plt.show()
```
